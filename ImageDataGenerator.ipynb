```
import tensorflow as tf
import numpy as np
```

```
img = tf.keras.preprocessing.image.load_img('/gdrive/My Drive/DeepLearning/images/Pinarello-Dogma.jpg')
x = tf.keras.preprocessing.image.img_to_array(img)
print(x.shape)
```

```
(1,) + x.shape # 이미지 갯수를 여러장 받아들이기 위해 차원을 확장함
```

```
x = x.reshape((1,) + x.shape)
```

```
data_aug_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255, 
                                                         rotation_range=15, #15도만큼 rotate
                                                         width_shift_range=0.1,
                                                         height_shift_range=0.1,
                                                         shear_range=0.5,
                                                         zoom_range=[0.8, 2.0],
                                                         horizontal_flip=True,
                                                         vertical_flip=True,
                                                         fill_mode='nearest')

imageGen = data_aug_gen.flow(x, batch_size=1)
```

```
import matplotlib.pyplot as plt
imgCnt = 0
for img in imageGen:
    imgCnt += 1
    if imgCnt < 10:
        plt.imshow(img.reshape((667, 1000, 3)))
        plt.show()
    else:
        break
```
